```python
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("https://maplestory.nexon.com/Guide/ProbabilityResult/MonthData")
soup = BeautifulSoup(target, "html.parser")
select_list = soup.select("#container")

for random in select_list:
    contents = random.find("div", class_="contents_wrap2 sm").get_text().strip()
    print(contents)
```
