```python
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("https://www.kr.playblackdesert.com/ko-KR/News/Notice?boardType=1")
soup = BeautifulSoup(target, "html.parser")

select_list = soup.select("#wrap > div > div.container > article > div.tab_container > div > div.thumb_nail_area > ul > li")


for notice in select_list:    
    #print(notice.find("span", class_="line_clamp").string)
    #print(notice.find("span", class_="desc").string)
    #print(notice.find("em", class_="thum_cate cate_noti").string)
    noti = notice.find("em", class_="thum_cate cate_noti").string
    clamp = notice.find("span", class_="line_clamp").string
    desc = notice.find("span", class_="desc").string
    print(noti, clamp, '\n', desc)
```

```python
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("https://www.kr.playblackdesert.com/ko-KR/News/Notice?boardType=2")
soup = BeautifulSoup(target, 'html.parser')
select_list = soup.select("#wrap > div > div.container > article > div.tab_container > div > div.thumb_nail_area > ul > li")

#print(select_list)
for update in select_list :    
    up = update.select_one("em").string
    clamp = update.select_one("span", class_="line_clamp").string
    desc = update.find("span", class_="desc").string
    print(up, clamp, '\n', desc)
```
